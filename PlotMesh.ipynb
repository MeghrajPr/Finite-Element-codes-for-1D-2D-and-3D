```python
from matplotlib.patches import Polygon
fig = plt.figure(figsize=(10,12))
ax = fig.add_subplot(111, facecolor='w', aspect='equal')

for i in range(eleconnect.shape[0]):
    if eleconnect[i,1] > 4:
        nnodcon = int(np.floor(eleconnect[i,1]/2)+1)        
    else:
        nnodcon = int(np.floor(eleconnect[i,1])+1)      
    ax.add_artist(Polygon(nodal_data[eleconnect[i,2:nnodcon+1].astype(int)-1,1:3],fc='c',ec='k',closed='True'))

ax.plot(nodal_data[:,1],nodal_data[:,2],'r.')

for i in range(nodal_data.shape[0]):
    ax.text(nodal_data[i,1]*1.02,nodal_data[i,2]*1.04, int(nodal_data[i,0]), horizontalalignment='center',verticalalignment='center', c='b')

ax.set_xlim(np.min(nodal_data[:,1]),np.max(nodal_data[:,1]))
ax.set_ylim(np.min(nodal_data[:,2]),np.max(nodal_data[:,2]))
#fig.tight_layout()
plt.show()
```
